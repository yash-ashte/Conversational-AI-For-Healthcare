Descriptive Functions

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
target_mapping = {i: str(iris.target_names[i]) for i in range(len(iris.target_names))}
data['species'] = data['species'].map(target_mapping)

data.head()
```

Model Based Functions

```python
def plot_distribution(feature, species=None, separate_species=False, plot_type='histogram'):
    
    df = data
    hue = None
    
    if species:
        df = df[df['species'] == species]

    if separate_species:
        hue = 'species'

    if plot_type == 'histogram':
        sns.histplot(data=df, x=feature, hue=hue, kde=True, palette='Set2')
        plt.title(f'Histogram of {feature} by Species')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
    elif plot_type == 'boxplot':
        sns.boxplot(x='species', y=feature, data=df, palette='Set2')
        plt.title(f'Box Plot of {feature} by Species')
    else:
        print("Invalid plot type. Choose either 'histogram' or 'boxplot'.")
    plt.show()

# Example usage:
plot_distribution('sepal width (cm)', plot_type='histogram', species='setosa')
plot_distribution('petal length (cm)', plot_type='histogram', separate_species=True)
plot_distribution('sepal width (cm)', plot_type='boxplot')
```
